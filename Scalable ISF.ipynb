```python
%matplotlib inline
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import exp

target = 100
isf = 80
```

```python
x = np.linspace(40,400,num=10)

# isf mg/dL/U
# scaling factor is mg/dL/U/mg/dL :)
isf_scaling_factor = 0.08
scaling_origin = 120

def isf_t(isf, bg):
    return isf + (bg-scaling_origin)*(-isf_scaling_factor)

y = np.array([isf for bg in x])
plt.plot(x, y, label="isf")

y = np.array([isf_t(isf,bg) for bg in x])
plt.plot(x, y, label="scaled isf (0.08)")

isf_scaling_factor = 0.1
y = np.array([isf_t(isf,bg) for bg in x])
plt.plot(x, y, label="scaled isf (0.1)")

plt.legend(loc='lower left')
plt.xlabel('glucose (mg/dL)')
plt.ylabel('sensitivity (mg/dL/U)')

```

```python

y = np.array([(bg-target)/isf for bg in x])
plt.plot(x, y, label="normal isf (80 mg/dL/U) correction ")

isf_scaling_factor = 0.08
y = np.array([(bg-target)/isf_t(isf,bg) for bg in x])
plt.plot(x, y, label="scaled isf (0.08 scaling factor) correction")

isf_scaling_factor = 0.1
y = np.array([(bg-target)/isf_t(isf,bg) for bg in x])
plt.plot(x, y, label="scaled isf (0.1 scaling factor) correction")

print(y)
plt.legend(loc='upper left')

plt.xlabel('glucose (mg/dL)')
plt.ylabel('insulin (U)')

```
